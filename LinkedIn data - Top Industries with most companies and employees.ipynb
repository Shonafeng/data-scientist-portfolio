```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv('temp_datalab_records_linkedin_company.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
#plot1: Count of Companies by top 20 industries as of most recent date
df_recent=df.loc[df.as_of_date==df.as_of_date.max(),:]
```

```python
grouped=df_recent.groupby(['industry'])['company_name'].count().reset_index()
group15=grouped.sort_values('company_name',ascending=False).head(15)

plt.figure(figsize=(10,6))
plt.bar(group15.industry,group15.company_name,align='center')
plt.title('TOP 15 Industries on LinkedIn with most companies as of 2018-07-17')
plt.xticks(rotation=90)
plt.show()
```

```python
grouped=df_recent.groupby(['industry'])['employees_on_platform'].sum().reset_index()
group15=grouped.sort_values('employees_on_platform',ascending=False).head(15)
plt.figure(figsize=(10,6))
plt.bar(group15.industry,group15.employees_on_platform,align='center')
plt.title('TOP 15 Industries on LinkedIn with most employees as of 2018-07-17')
plt.xticks(rotation=90)
plt.show()
```
